```python
def make_family_bins(family_size):
    if family_size == 1:
        return 'Alone'
    if family_size <= 3:
        return 'Small'
    if family_size <= 4:
        return 'Kernel'
    else: return 'Large'
```

```python
train['FamilyBins'] = train["FamilySize"].apply(make_family_bins)
```
